# One_hot_encoding

```python
import numpy as np
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sentences = ['i am jitesh sahu', 'i am a mechanical engineering student', 'i belong from bargarh', 
             'i am a fresher', 'playing badminton is my hobby', 'there is a long way to go']
```

```python
## vocabulary size = 10000
voc_size = 10000
```

# One Hot Representation

```python
onehot_repre = [one_hot(word, voc_size) for word in sentences]

print(onehot_repre)
```

# Word Embedding Represntation

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
sent_length = 10
embedded_docs = pad_sequences(onehot_repre, padding='pre', maxlen = sent_length) 

print(embedded_docs)                              
```

```python
length = 10
docs = pad_sequences(onehot_repre, padding='post', maxlen = length)

print(docs)
```

```python
dim = 0
```

```python
model = Sequential()
model.add(Embedding(voc_size, 10, input_length= sent_length))
model.compile('adam', 'mse')
```

```python
model.summary()
```

```python
print(model.predict(embedded_docs))
```
