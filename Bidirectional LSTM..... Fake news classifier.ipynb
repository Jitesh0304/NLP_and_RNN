```python
import pandas as pd
import numpy as np
import nltk
import re
from nltk.corpus import stopwords
```

```python
df = pd.read_csv('fake_news_train.csv')

df.head()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.shape
```

```python
df = df.reset_index()
df.head()
```

```python
df = df.drop(['index','id'], axis = 1)
df.head()
```

```python
x = df.drop('label', axis =1)

y = df['label']
```

```python
x.shape
```

```python
y.shape
```

```python
message = x.copy()

message
```

```python
from tensorflow.keras.layers import Embedding                            ## embedding layer
from tensorflow.keras.preprocessing.sequence import pad_sequences

from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot      ## one hot encoding

from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense

from tensorflow.keras.layers import Bidirectional
from tensorflow.keras.layers import Dropout
```

```python
from nltk.stem.porter import PorterStemmer

stemmer = PorterStemmer()
```

```python
corpus = []

for i in range(0, len(message)):
    
    review = re.sub('[^a-zA-Z]',' ', message['title'][i])
    
    review = review.lower()
    
    review = review.split()
    
    review = [stemmer.stem(word) for word in review if word not in stopwords.words('english')]
    
    review = ' '.join(review)
    
    corpus.append(review)
```

```python
corpus[6]
```

```python
### Vocabulary size
voc_size=5000
```

```python
onehot_repre = [one_hot(words, voc_size) for words in corpus]

onehot_repre
```

# Embedding Representation

```python
sent_length = 20              

embedded_docs = pad_sequences(onehot_repre, padding= 'pre', maxlen = sent_length)     
                                     
print(embedded_docs)
```

## model 1........ LSTM model

```python
embedding_vector_features = 40

model = Sequential()

model.add(Embedding(voc_size, embedding_vector_features, input_length= sent_length))

model.add(LSTM(100)) 

model.add(Dense(1, activation= 'sigmoid'))     
```

```python
model.compile( loss= 'binary_crossentropy', optimizer ='adam', metrics= ['accuracy'])

print(model.summary())
```

# model 2 ..... Bidirectional LSTM

```python
embedding_vector_features = 40

model_2 = Sequential()

model_2.add(Embedding(voc_size, embedding_vector_features, input_length= sent_length))

model_2.add(Bidirectional(LSTM(100)))  

model_2.add(Dropout(0.3))

model_2.add(Dense(1, activation= 'sigmoid'))   
```

```python
model_2.compile(loss ='binary_crossentropy', optimizer ='adam', metrics =['accuracy'])

print(model_2.summary())
```

```python
## convert the embedded documents to an arry

x_final = np.array(embedded_docs)

y_final = np.array(y)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_final, y_final, random_state= 0, test_size= 0.33)
```

```python
model.fit(x_train, y_train, validation_data= (x_test, y_test), epochs= 10, batch_size =64)
```

```python
y_pred = model.predict(x_test)
```

```python
pred2 = (y_pred > 0.5).astype("int32")
```

```python
pred2
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
confusion_matrix(y_test, pred2)
```

```python
accuracy_score(y_test, pred2)
```
