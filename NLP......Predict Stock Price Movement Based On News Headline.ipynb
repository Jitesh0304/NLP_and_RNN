```python
import pandas as pd
```

```python
df = pd.read_csv('Data.csv',  encoding = "ISO-8859-1")

df.head(3)
```

```python
df.shape
```

```python
train = df[df['Date'] < '20150101']

test = df[df['Date'] > '20141231']
```

```python
train.shape
```

```python
test.shape
```

```python
## removing punctuation
data = train.iloc[:,2:27]

data.replace("[^a-zA-Z]"," ", regex =True, inplace = True)
```

```python
data.head(3)
```

```python
## change all columns name

list1 = [i for i in range(25)]
new_index = [str(i) for i in list1]

data.columns = new_index
data.head(3)
```

```python
for i in new_index:
    data[i] = data[i].str.lower()
    
data.head(1)
```

```python
' '.join(str(x) for x in data.iloc[1,0:25])   
```

```python
data.index
```

```python
## capturing all headline and join them then only we can use (count_vectorizer  or  TF-IDF)

headline =[]
for i in range(0,len(data.index)):
    headline.append(' '.join(str(x) for x in data.iloc[i,0:25]))
```

```python
headline[0]
```

```python
## import all required libraries

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
## implement bag of words

cv = CountVectorizer(ngram_range= (2,2))

train_dataset = cv.fit_transform(headline)
```

```python
train_dataset
```

```python
## implement random forest_classifier

rf_classifier = RandomForestClassifier(n_estimators= 200, criterion= 'entropy')
```

```python
rf_classifier.fit(train_dataset, train['Label'])           ## x => train_dataset      y => Label column of the dataset
```

```python
## predict for the test dataset

test_transform = []
for i in range(0,len(test.index)):
    test_transform.append(' '.join(str(x) for x in test.iloc[i,2:27]))
    
test_dataset = cv.transform(test_transform)
```

```python
test_dataset[0]
```

```python
prediction = rf_classifier.predict(test_dataset)
```

```python
matrics = confusion_matrix(prediction, test['Label'])
matrics
```

```python
score = accuracy_score(test['Label'], prediction)
score
```

```python
report = classification_report(prediction, test['Label'])
print(report)
```
