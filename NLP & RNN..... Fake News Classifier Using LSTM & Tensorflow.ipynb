```python
y = df['label']
y.head()
```

```python
message = df.copy()

message.shape
```

```python
message.reset_index(inplace= True)
message.head()
```

```python
message['title'][0]
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

```python
stemmer = PorterStemmer()
```

```python
corpus =[]

for i in range(0, len(message)):
    
    review = re.sub('[^a-zA-Z]',' ',message['title'][i])
    
    review = review.lower()
    
    review = review.split()
    
    review = [stemmer.stem(word) for word in review if word not in set(stopwords.words('english'))]
    
    review = ' '.join(review)
    
    corpus.append(review)
```

```python
corpus[3]
```

# Onehot Representation

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
```

```python
## set the vocabulary size
voc_size = 5000
```

```python
onehot_repre = [one_hot(word, voc_size) for word in corpus]

onehot_repre
```

# Embedding Representation

```python
sent_length = 20

embedded_docs = pad_sequences(onehot_repre, padding= 'pre', maxlen = sent_length)

embedded_docs
```

```python
embedded_docs[0]
```

# creating model

```python
embedding_vector_features = 40

model =Sequential()

model.add(Embedding( voc_size, embedding_vector_features, input_length= sent_length))

model.add(LSTM(100))

model.add(Dense(1 , activation= 'sigmoid'))

model.compile( loss= 'binary_crossentropy', optimizer = 'adam', metrics= ['accuracy'])

model.summary()
```

```python
len(embedded_docs)
```

```python
y.shape
```

```python
import numpy as np

x_final = np.array(embedded_docs)
y_final = np.array(y)
```

```python
x_final.shape
```

```python
x.shape
```

```python
y_final.shape
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_final, y_final, random_state= 0, test_size= 0.33)
```

# Model Training

```python
model.fit(x_train, y_train, validation_data= (x_test, y_test), epochs= 10, batch_size = 65)
```

# Performance Metrics And Accuracy

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
y_pred = model.predict(x_test)
```

### Prediction_1 (how to predict the class)

```python
classes_x=np.argmax(y_pred,axis=1)
```

```python
classes_x
```

```python
confusion_matrix(y_test, classes_x)
```

```python
accuracy_score(y_test, classes_x)
```

### Prediction_2 (how to predict the class)

```python
pred2 = (y_pred > 0.5).astype("int32")
```

```python
pred2
```

```python
confusion_matrix(y_test, pred2)
```

```python
accuracy_score(y_test, pred2)
```

# Adding Dropout

```python
from tensorflow.keras.layers import Dropout
```

```python
embedding_vector_features = 40

model =Sequential()

model.add(Embedding( voc_size, embedding_vector_features, input_length= sent_length))

model.add(Dropout(0.3))

model.add(LSTM(100))

model.add(Dropout(0.3))

model.add(Dense(1 , activation= 'sigmoid'))

model.compile( loss= 'binary_crossentropy', optimizer = 'adam', metrics= ['accuracy'])

model.summary()
```

```python
import numpy as np

x_final = np.array(embedded_docs)
y_final = np.array(y)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_final, y_final, random_state= 0, test_size= 0.33)
```

```python
model.fit(x_train, y_train, validation_data= (x_test, y_test), epochs= 10, batch_size = 65)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
y_pred = model.predict(x_test)
```

```python
pred1 = (y_pred > 0.5).astype("int32")
```
